```python
import pandas as pd

base_path = './data/Stage/'
target_path = './data/Target/'

customer_df = pd.read_csv(base_path + 'customer.csv')
sales_df = pd.read_csv(base_path + 'sales.csv')
```

```python
customer_df.columns
```

```python
### target to state dimension
state = customer_df[['State']]
### target to city dimension
city = customer_df[['State', 'City']]
### target to customer dimension
```

```python
state.drop_duplicates(inplace=True)
state.reset_index(drop=True, inplace=True)
state.loc[:, ['StateKey']] = state.index + 1
state
```

```python
city_rename = {'City': 'CityName'}
city_cols = ['CityKey', 'StateKey', 'CityName']
city.drop_duplicates(inplace=True)
city = city.join(state.set_index('State'), on='State')
city.reset_index(drop=True, inplace=True)
city['CityKey'] = city.index + 1
city.rename(columns=city_rename, inplace=True)
city = city[city_cols]
city
```

```python
city['CityName'].value_counts()
```

```python
state.to_csv(target_path + 'DimState.csv', index=False)
city.to_csv(target_path + 'DimCity.csv', index=False)
```
