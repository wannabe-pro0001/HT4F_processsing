```python
import pandas as pd

def generate_date_df(start=pd.Timestamp('2021-01-01 00:00:00'), end=pd.Timestamp('2021-12-31 00:00:00')):
    df = pd.DataFrame(pd.date_range(start, end), columns=['Date'])
    # df = pd.DataFrame(pd.date_range(start, end), columns=['Date'])
    df['DateKey'] = df['Date'].dt.strftime('%Y%m%d')
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['WeekDayName'] = df['Date'].dt.day_name()

    return df

df = generate_date_df(pd.Timestamp('2022-10-01 00:00:00'), pd.Timestamp('2022-12-31 00:00:00'))
df
```

```python
df.to_csv('Data/Target/DimDate.csv', index=False)
```
